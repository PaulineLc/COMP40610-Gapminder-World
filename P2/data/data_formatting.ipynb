```python
import pandas as pd
import numpy as np
```

```python
fertility_df = pd.read_csv('fertility.csv', skiprows=4)
```

```python
agriculture_df = pd.read_csv('agriculture.csv', skiprows=4)
```

```python
agriculture_df.tail()
```

```python
gdp_df = pd.read_csv('gdp.csv', skiprows=4)
```

```python
gdp_df.head()
```

```python
df = pd.DataFrame({'Country': [], 'Code': [], 'Year': [], 'GDP': [], 'Fertility':[], 'Agriculture': []}, 
                  columns=['Country', 'Code', 'Year', 'GDP', 'Fertility', 'Agriculture'])
```

```python
for i in range(fertility_df.shape[0]):
    country = fertility_df['Country Name'][i]
    code = fertility_df['Country Code'][i]
    for j in range(1960, 2016+1):
        year = j
        agriculture = float(agriculture_df[str(year)].loc[agriculture_df['Country Code'] == code])
        fertility = float(fertility_df[str(year)].loc[fertility_df['Country Code'] == code])
        gdp = float(gdp_df[str(year)].loc[gdp_df['Country Code'] == code])
        if np.isnan(agriculture) or np.isnan(fertility) or np.isnan(gdp):
            pass
        else:
            df.loc[-1] = [country, code, year, gdp, fertility, agriculture]  # adding a row
            df.index = df.index + 1  # shifting index

df = df.sort_index()  # sorting by index
```

```python
df.to_csv('consolidated_data.csv')
```
